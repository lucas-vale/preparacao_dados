```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('clientes-v2-tratados.csv')
```

```python
df.head()
```

```python
# Codificando one-hot para 'estado_civil

df = pd.concat([df, pd.get_dummies(df['estado_civil'], prefix = 'estado_civil')], axis = 1)
```

```python
df.head()
```

```python
# Codificacao ordinal para 'nivel_ educacao

educacao_ordem = {'Ensino Fundamental': 1, 'Ensino Médio': 2, 'Ensino Superior': 3, 'Pós-Graduação': 4}
df['nivel_educaco_ordinal'] = df['nivel_educacao'].map(educacao_ordem)
```

```python
df.head()
```

```python
# Transformar 'area_atuacao' em categorias codificadas usando metodo cat.codes

df['area_atuacao_cod'] = df['area_atuacao'].astype('category').cat.codes
```

```python
df.head()
```

```python
label_encoder = LabelEncoder()
df['estado_cod'] = label_encoder.fit_transform(df['estado'])
```
